```python
import random
import string

# ANSI escape codes for color
RED = "\033[91m"
RESET = "\033[0m"

def generate_password(length):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

# Generate and print passwords from 1 to 40 characters with color coding
for length in range(1, 41):
    password = generate_password(length)
    color = RED if length < 10 else RESET
    print(f"{color}Password of length {length}: {password}{RESET}")
```

```python
import random
import string
import math
import pyperclip
import nltk

import os
os.environ["NLTK_DATA"] = "M:\\nltk_data"
from nltk.corpus import words


# Ensure you have the word list
nltk.download("words", download_dir="M:\\nltk_data")
word_list = words.words()

# ANSI escape codes for color
RED = "\033[91m"
GREEN = "\033[92m"
RESET = "\033[0m"

def calculate_entropy(password):
    """Estimates password strength based on entropy calculation."""
    character_set_size = len(set(password))  # Unique characters in the password
    entropy = math.log2(character_set_size ** len(password))
    return entropy

def generate_password(length):
    """Generates a random password with the given length."""
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    entropy = calculate_entropy(password)
    return password, entropy

def generate_passphrase(num_words=4):
    """Generates a passphrase using random words."""
    passphrase = '-'.join(random.choice(word_list) for _ in range(num_words))
    entropy = calculate_entropy(passphrase)
    return passphrase, entropy

# User input for password type
choice = input("Generate (P)assword or (W)ord-based Passphrase? ").strip().lower()
length_or_words = int(input("Enter the number of characters/words: "))

if choice == 'p':
    password, entropy = generate_password(length_or_words)
elif choice == 'w':
    password, entropy = generate_passphrase(length_or_words)
else:
    print("Invalid choice. Please restart.")
    exit()

# Determine color based on entropy strength
color = RED if entropy < 40 else GREEN
print(f"{color}Generated: {password} (Entropy: {entropy:.2f}){RESET}")

# Copy to clipboard
pyperclip.copy(password)
```
